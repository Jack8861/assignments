```python
from bs4 import BeautifulSoup 
 
import requests
 
import nltk 
 
import pprint
```

```python
url = 'http://php.net/'
```

```python
# html = response.read() 
res = requests.get(url)
html_page = res.content
```

```python
soup = BeautifulSoup(html_page, 'html.parser')
```

### In this assignment students have to find the frequency of words in a webpage.

### so... ill just use the bag of words concept to get the frequency of words

```python
from sklearn.feature_extraction.text import CountVectorizer 
from nltk.tokenize import word_tokenize

string = soup.find_all(text=True)
vect1 = CountVectorizer()
vect_fit = vect1.fit_transform(string)

# obtain the list of words...
word_list = vect1.get_feature_names();   
# obtain the corresponding list of "no of words"
count_list = vect_fit.toarray().sum(axis=0)

# make it into a dictionary
pprint.pprint(dict(zip(word_list,count_list)))
```
