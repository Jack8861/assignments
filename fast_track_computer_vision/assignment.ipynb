2. Create the convolutional kernel matrix for a top edge detector.

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from skimage import io

import cv2
import numpy as np

%matplotlib inline

# Read in the image
image = mpimg.imread("/content/download.jpeg")

plt.imshow(image)

gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)


# top edges detected
sobel_x = np.array([[ 1, 2, 1], 
                   [ 0, 0, 0], 
                   [ -1, -2, -1]])


filtered_image1 = cv2.filter2D(gray, -1, sobel_x)
f, ax = plt.subplots(1, figsize=(15, 4))
ax.set_title('horizontal edge detection')
ax.imshow(filtered_image1, cmap='gray')
```
